# Curva de Bézier: Genérica
image.png

```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

grau_curva = int(input("Insira qual o grau da curva desejado: "))

ponto_controle = np.random.rand(grau_curva + 1, 2) * 10

t = np.linspace(0, 1, 100).reshape(-1, 1) # Cria 100 valores de t entre 0 e 1 → usados para gerar a curva.

curva = 0

# soma das bases de Bernstein
for k in range(grau_curva + 1):
  curva += comb(grau_curva, k) * (1 - t)**(grau_curva - k) * t**k * ponto_controle[k]

plt.plot(curva[:, 0], curva[:, 1], 'b-', label='Curva de Bézier')
plt.plot(ponto_controle[:, 0], ponto_controle[:, 1], 'ro--', label='Pontos de Controle')
for i, (x, y) in enumerate(ponto_controle):
  plt.text(x+0.2, y+0.2, f'A{i}')
plt.title(f'Curva de Bézier (grau {grau_curva})')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```
